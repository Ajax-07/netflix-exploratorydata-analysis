## Asking and Answering Questions

visualizing data 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
from collections import Counter 
```

### Q.  What is release groth over year

```python
matplotlib.rcParams['font.size']=18;
plt.figure(figsize=(10,8))
plt.plot(release_over_A_year,'-g',lw=4,)
plt.title("release growth per Year")
plt.xlabel("Release count")
plt.ylabel("years")
```

### Q. how many variety  of shows and movies and which is popular

```python
tp = netflix_df[netflix_df['listed_in'].notnull()]

all_type = np.array(tp.listed_in)
```

```python
type_df = []
for t in all_type:
    x = list(t.split(', '))
    for i in x:
        type_df.append(i)
        
all_type_df = list(dict.fromkeys(type_df))
```

```python
all_type_df
```

```python
print(f"Total {len(all_type_df)} variety  of shows and movies are avaliable ")
```

```python
CountOfvariety  = dict(Counter(type_df))
```

```python
CountOfvariety
```

```python
key = []
val = []
for k in CountOfvariety.keys():
    key.append(k)
    
for v in CountOfvariety.values():
    val.append(v) 
```

```python
plt.figure(figsize=(20,8))
plt.plot(range(0,42),val,'o-y',lw=4, ms=10, mew=2, mec='black',mfc='blue')
plt.title('Number of content variety graph')
plt.xlabel('index of key list of name of variety')
plt.ylabel('count of Variety')
```

```python
#Number of count of variety in another form
plt.figure(figsize=(110,30))
sns.countplot(type_df)
plt.title('popularity of each Variety content')
```

### Q. How was data over india

### Q. What is type of rating on each content

```python
rate_df = netflix_df.groupby('rating')['title'].count().to_frame()
rate_df.style.background_gradient(cmap = "Reds")
```
